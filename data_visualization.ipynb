### Task 1: Exploring and Visualizing Dataset
#### Problem Statement:
    The goal is to explore and visualize the Iris dataset to gain insights into the data.
#### Goal:
    The goal is to perform the exploratory data Analysis(EDA) to undertsand its main characteristics.

```python
# Import Libraries
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(style= "whitegrid")
```

### Load Dataset

```python
#Loading Iris dataset from sklearn library them making its dataframes using pandas
#and printing its first few(5) rows
iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

print(df.head())
```

### Basic Information about dataset

```python
#Printing basic information related to dataset like its shape(dimensions), 
#col names, info(dattypes missing values), describe(mean, med, std)
print("Dataset Shape: ", df.shape)
print("Column Names: ", df.columns.tolist())
print("Dataset Info: ", df.info())
print("Summary Statistics:", df.describe())
```

### Scatter Plot

```python
#Scatter plot using seaborn. it shows how the features are related to eachother
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="sepal length (cm)", y="sepal width (cm)", hue="species")
plt.title("Sepal Length vs Width")
plt.show()
```

### Histogram

```python
#Histogram of all numeric columns. Its hows ar values are distributed
df.hist(figsize=(10,8))
plt.suptitle("Feature Distributions", size=16)
plt.show()
```

### Box Plot

```python
#Box plot help us to detect outliers. It also show med, quartiles, range
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Box Plot of Iris Features")
plt.show()
```

### PairPlot

```python
#Pairplot help us to discover patterns. It show all combinations of Scatter or histogram plot
sns.pairplot(df, hue="species")
plt.show()
```

### Final Insights

        Setosa species generally has smaller petal sizes
        Virginica species has the largest petal size
        Scatter plots show clear separation between species
        Box plots reveal outliers in the dataset
